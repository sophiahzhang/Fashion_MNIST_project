### Basic regression model with Regularization

```python
#Load data
X = np.loadtxt('./data_shirts/train_shirt_x.csv', delimiter=',', skiprows=1)
y = np.loadtxt('./data_shirts/train_shirt_y.csv', delimiter=',', skiprows=1)

#Split the data
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#Use max iterations = 100 for model fit (default)
C_grid = np.logspace(-9, 6, 31)
#C=inverse of regularization strength 
#i.e. higher C, lower regularization (less penalty on complex models) and more overfitting/accuracy on training
min_logloss=100
accuracy_at_min=0
C_at_min=0
cm=np.zeros([2,2])
coef=np.zeros([1, 784])
    #initialize lists
scores=[]
loglosses=[]
predictions=[]
for C in C_grid:
# Build and evaluate model for each value C
#‘auto’ selects ‘ovr’ (one vs rest) if the data is binary, or if solver=’liblinear’,
    model = LogisticRegression(C=C, solver='liblinear',max_iter=100)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    score = model.score(x_test, y_test)
    logloss = log_loss(y_test, y_pred)
    scores.append(score)
    loglosses.append(logloss)
    if logloss < min_logloss or (logloss == min_logloss and C > C_at_min): 
        min_logloss=logloss 
        C_at_min=C
        accuracy_at_min=score
        cm = sklearn.metrics.confusion_matrix(y_test, y_pred)
        predictions=y_pred
        coef=model.coef_
```

#### Find Best N for PCA

```python
#PCA

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = np.loadtxt('./data_shirts/train_shirt_x.csv', delimiter=',', skiprows=1)
y = np.loadtxt('./data_shirts/train_shirt_y.csv', delimiter=',', skiprows=1)

X_c=X.copy()

N_components = np.arange(2, 200, 5)

scores=[]
loglosses=[]
for n in N_components:
    pca = PCA(n_components = n) # n components is the number of components you want to reduce the dimensions to 
    pca.fit(X_c) #Apply dimensionality reduction to scaled_data
    X = pca.transform(X_c)
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    model = LogisticRegression(C=0.1, solver='liblinear',max_iter=100)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    score = model.score(x_test, y_test)
    logloss = log_loss(y_test, y_pred)
    scores.append(score)
    loglosses.append(logloss)
```

```python
#Load the data
X = np.loadtxt('./data_shirts/train_shirt_x.csv', delimiter=',', skiprows=1)
y = np.loadtxt('./data_shirts/train_shirt_y.csv', delimiter=',', skiprows=1)

X_c=X.copy()

N_components = 97
pca = PCA(N_components)
pca.fit(X_c)
X = pca.transform(X_c)

#Split the data
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#Use max iterations = 100 for model fit (default)
C_grid = np.logspace(-9, 6, 31)
#C=inverse of regularization strength 
#i.e. higher C, lower regularization (less penalty on complex models) and more overfitting/accuracy on training
min_logloss=100
accuracy_at_min=0
C_at_min=0
cm=np.zeros([2,2])
coef=np.zeros([1, 784])
    #initialize lists
trscores=[]
trloglosses=[]
scores=[]
loglosses=[]
predictions=[]
for C in C_grid:
# Build and evaluate model for each value C
#‘auto’ selects ‘ovr’ (one vs rest) if the data is binary, or if solver=’liblinear’,
    model = LogisticRegression(C=C, solver='liblinear',max_iter=100)
    model.fit(x_train, y_train)
    y_trpred = model.predict(x_train)
    y_pred = model.predict(x_test)
    score = model.score(x_test, y_test)
    logloss = log_loss(y_test, y_pred)
    scores.append(score)
    loglosses.append(logloss)
    trscore = model.score(x_train, y_train)
    trlogloss = log_loss(y_train, y_trpred)
    trscores.append(trscore)
    trloglosses.append(trlogloss)
    if logloss < min_logloss or (logloss == min_logloss and C < C_at_min): 
        min_logloss=logloss 
        C_at_min=C
        accuracy_at_min=score
        cm = sklearn.metrics.confusion_matrix(y_test, y_pred)
        predictions=y_pred
        coef=model.coef_
```

### Squaring the pixels

```python
#Squaring the data
from sklearn.base import BaseEstimator, TransformerMixin
class SquareFeatureExtractor(BaseEstimator, TransformerMixin):
    """ Squares each feature in the input vector
    """

    def fit(self, x_NF, y_N=None):
        """ Nothing happens, as this transform has no trainable parameters.
        """
        return self
    
    def transform(self, x_NF, y_N=None):
        """ 
        Squares each feature in original data
        Returns
        -------
        feat_NF : 2D array, size N x F
            F features extracted for each example
        """
        return np.square(x_NF)
```

For the values i = 1, 2, . . . , 40, build a logistic regression model with the max_iter set to i. Fit each such model to the training data, and keep track of the accuracy of the resulting model (via the model’s own score() function) along with the logistic loss, each measured on the training data.† <br>
Produce two plots, each with the values of i as x-axis and with the accuracy/loss, respectively, as y. Place these plots into your PDF document, with captions labeling each appropriately. Below the plots, discuss the results you are seeing; what do they show, and why? <br>

When doing this, you will probably see warnings about non-convergence for lower values of parameter i. You can ignore these warnings, as they are expected. You can measure the loss using 
    [log_loss](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.log_loss.html)
    <br>
sklearn.metrics.log_loss(y_true, y_pred, *, eps=1e-15, normalize=True, sample_weight=None, labels=None)

sklearn.linear_model.LogisticRegression(penalty='l2', *, dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, solver='lbfgs', max_iter=100, multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None)

attributes: <br>

coef_ is of shape (1, n_features) when the given problem is binary. In particular, when multi_class='multinomial', coef_ corresponds to outcome 1 (True) and -coef_ corresponds to outcome 0 (False).

```python
#ignore warnings
trscores=[]
scores=[]
trloglosses=[]
loglosses=[]

#‘auto’ selects ‘ovr’ (one vs rest) if the data is binary, or if solver=’liblinear’,
#inverse of regularization strength
for i in range(1,41):
    model = LogisticRegression(solver='liblinear',max_iter=i)
    model.fit(x_train, y_train)
    ytr_pred = model.predict(x_train)
    y_pred = model.predict(x_test)
    tr_score = model.score(x_train, y_train)
    score = model.score(x_test, y_test)
    trlogloss = log_loss(y_train, ytr_pred)
    logloss = log_loss(y_test, y_pred)
    trscores.append(tr_score)
    scores.append(score)
    loglosses.append(logloss)
    trloglosses.append(trlogloss)

#print("Basic accuracy: ", score)
#print("Log loss: ", score)
```

Produce two plots, each with the values of i as x-axis and with the accuracy/loss, respectively, as y. Place these plots into your PDF document, with captions labeling each appropriately. Below the plots, discuss the results you are seeing; what do they show, and why? 

```python
# plot lines for training and test data comparison for accuracy
plt.plot(np.arange(1,41), trscores, color='red',label="Training Data")
plt.plot(np.arange(1,41), scores, color='blue',label="Testing Data")
plt.title('Training and Test Data Accuracy vs Max No. Iterations')
plt.xlabel('Max No. Iterations');
plt.ylabel('Accuracy');
plt.legend()
plt.show()
```

```python
# plot lines for training and test data comparison for log loss
plt.plot(np.arange(1,41), trloglosses, color='red',label="Training Data")
plt.plot(np.arange(1,41), loglosses, color='blue',label="Testing Data")
plt.title('Training and Test Data Log Loss vs Max No. Iterations')
plt.xlabel('Max No. Iterations');
plt.ylabel('Log-loss');
plt.legend()
plt.show()
```

When we limit the iterations allowed for the solver to converge on its solution, we limit the number of updates to the model weights. The more iterations we have the more the weights can be updated to those that generate the best fit model. As you can see in figure 1, the training accuracy increases dramatically from 1 to 8 iterations. The algorithm nudges the weights in the right direction to the optimal set of weights. There is no significant change in weights or accuracy after 10 iterations because it has converged. The accuracy on the testing data shows a similar trend, the accuracy is changing due to fluctuations in the weights of the fitted model at lower max iterations. A minimum number of iterations is needed for the solver to converge, and in this case it the number appears to be 10. The algorithm uses a 'one-vs-rest' method to separate the two binary classes, it needs several passes to check the separator works. In this case the testing accuracy actually decreases from 1 to later iterations but remains steady beyond 10 iterations. Similarly, the training log-loss of the model also show a sharp decrease from 1 to 8 iterations and the process converges to the best weights after that. The log-loss increases from initial value of 1.22 to a stable value around 1.30 at algorithm convergence. This is aligned with the decrease in accuracy, since the error term in the loss function increases.

```python
#Use max iterations = 100 for model fit (default)
C_grid = np.logspace(-9, 6, 31)
#C=inverse of regularization strength 
#i.e. higher C, lower regularization (less penalty on complex models) and more overfitting/accuracy on training
min_logloss=100
accuracy_at_min=0
C_at_min=0
cm=np.zeros([2,2])
coef=np.zeros([1, 784])
    #initialize lists
scores=[]
loglosses=[]
predictions=[]
for C in C_grid:
# Build and evaluate model for each value C
#‘auto’ selects ‘ovr’ (one vs rest) if the data is binary, or if solver=’liblinear’,
    model = LogisticRegression(C=C, solver='liblinear',max_iter=100)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    score = model.score(x_test, y_test)
    logloss = log_loss(y_test, y_pred)
    scores.append(score)
    loglosses.append(logloss)
    if logloss < min_logloss or (logloss == min_logloss and C > C_at_min): 
        min_logloss=logloss 
        C_at_min=C
        accuracy_at_min=score
        cm = sklearn.metrics.confusion_matrix(y_test, y_pred)
        predictions=y_pred
        coef=model.coef_
```
